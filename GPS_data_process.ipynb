Import libraries

```python
import pandas as pd
import datetime
from datetime import datetime, timedelta
import os
import matplotlib.pyplot as plt
```

Import data

```python
read_data = pd.read_csv("/content/datafarmGPS.csv")
print(read_data) 
```

```python
data_list=list(read_data) #convert dataframe into list for easier splitting
```

```python
splt1 = data_list[3].split(", ") #split the different parameters from each other
splt1
```

```python
splt2 = splt1[1].split("=") #split again to be able to extract only the values of the parameters
splt2[1]
```

Extract only the values of the wanted parameters:
Time
Latitude
Longitude
Quality
NumSV
Altitude




```python
wanted_params=[1,2,4,6,7,9]
inlist= []
outlist= []
for i in range(len(data_list)):
  splt1 = data_list[i].split("', ")
  for j in wanted_params:
    splt2 = splt1[j].split("=")
    inlist.append(splt2[1])
  outlist.append(inlist)
  inlist= []
```

```python
print(outlist)
```

```python
# Create dataframe of the wanted values
df = pd.DataFrame(outlist, columns=['Time','Latitude','Longitude','Quality','Number of satellites','Altitude'])
for i in range(len(data_list)):
  df["Time"][i]= df["Time"][i][:-1]

df
```

Add 2 hours to the time
(because the timezone of the GPS is 2 hours behind)

```python
for i in range(len(data_list)):
  time_obj = datetime.strptime(df["Time"][i], "%H:%M:%S")
  time_obj += timedelta(hours=2)
  df["Time"][i] = time_obj.time()

df
```

```python
df.to_csv(r'LemonGPSlocation1.csv')
```
